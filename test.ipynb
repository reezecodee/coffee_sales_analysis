```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Data contoh
labels1 = ['A', 'B', 'C']
values1 = [30, 40, 30]

labels2 = ['X', 'Y', 'Z']
values2 = [20, 50, 30]

# Membuat subplot dengan dua pie chart menggunakan Plotly Express
fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]], subplot_titles=('Pie Chart 1', 'Pie Chart 2'))

fig.add_trace(go.Pie(labels=labels1, values=values1), 1, 1)
fig.add_trace(go.Pie(labels=labels2, values=values2), 1, 2)

# Menambahkan judul
fig.update_layout(title_text="Contoh Subplot Pie Chart")

# Menampilkan subplot
fig.show()
```
